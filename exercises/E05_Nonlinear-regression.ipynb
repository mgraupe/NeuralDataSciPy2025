#### Loading and visualiztion of the data

The wage data is saved in the numpy file `Wages.npy`. We start out by loading the file (using `np.load()`) and visualizing the data. 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

%matplotlib inline

data = np.load('Wages.npy')
print(np.shape(data))
data_x = data[:,0]
data_y = data[:,1]

plt.scatter(data_x, data_y, facecolor='None', edgecolor='k', alpha=0.3)
plt.xlabel('age (years)',fontsize=16)
plt.ylabel('yearly wage (k $)',fontsize=16)
plt.show()
```

Scikit-learn provides a function for exactly that purpose, i.e., splitting the data into a traning and a test subset : `train_test_split(x,y,test_size=[fraction],random_state=[integer])`. Here, we will use 66 % of the data-points for traning and 33 % for testing, i.e., `test_size=0.33` in `train_test_split()`.

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.33, random_state = 1)
```
